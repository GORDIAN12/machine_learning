```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('ingreso.csv')
df.head()
```

```python
df.plot.scatter(x='horas', y='ingreso', color='green')
```

```python
model=LinearRegression(fit_intercept=True)
```

```python
x=df[['horas']]
y=df['ingreso']
x.head()
```

```python
model.fit(x,y)
```

```python
#Pendiente
model.coef_
```

```python
#INterseccion
model.intercept_
```

```python
#salario estimado para X horas por cada persona
input=pd.DataFrame({'horas': [39,40,43,44]})
model.predict(input)
```

```python
plt.scatter(input,model.predict(input),color='red')
plt.plot(input,model.predict(input),color='black')

plt.ylabel('ingreso')
plt.xlabel('horas')
plt.scatter(df['horas'],df['ingreso'],color='blue')
plt.show()
```
